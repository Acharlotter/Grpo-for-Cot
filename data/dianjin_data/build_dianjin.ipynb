```python
# build_qwen_dataset.py
from datasets import load_dataset
import json
import os

# 1) 下载原始数据集
raw_ds = load_dataset("DianJin/DianJin-R1-Data", split="train")

# 2) 清洗成 Qwen 格式
```

```python
raw_ds[0]
```

```python
qwen_data = []
for sample in raw_ds:
    # 过滤空字符串
    #system = sample.get("system", "").strip()
    user     = sample.get("instruction", "").strip()
    assistant= sample.get("output", "").strip()
    if not (user and assistant):
        continue

    conversations = []
    # if system:                           # system 可选
    #     conversations.append({"from": "system",  "value": system})
    conversations.append({"from": "user",      "value": user})
    conversations.append({"from": "assistant", "value": assistant})
    qwen_data.append({"conversations": conversations})
```

```python
# 3) 保存为 .jsonl
output_path = "qwen_finance.jsonl"
with open(output_path, "w", encoding="utf-8") as f:
    for item in qwen_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print(f"✅ 已生成 {len(qwen_data)} 条样本 → {os.path.abspath(output_path)}")
```
